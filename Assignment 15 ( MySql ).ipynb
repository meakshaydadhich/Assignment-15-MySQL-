```python
# Q.1 What is database ? Differentiate between SQL and NoSQL databases.

# Ans.
# A database is an organized collection of data, typically stored and accessed electronically. Databases allow for the efficient
# retrieval, insertion, and management of large amounts of data. Databases can be managed using a Database Management System 
# (DBMS), which provides tools to create, read, update, and delete data.

# SQL databases are ideal for structured data with complex queries and strong consistency requirements.

# NoSQL databases are suitable for handling large-scale, unstructured data with flexible schemas and faster development speed.
```

```python
# Q.2 What is DDl ? Expalin why CREATE, DROP, ALTER and TRUNCATE are used with an example.

# Ans.
# DDL (Data Definition Language) is a subset of SQL used to define, alter, and manage database structures. DDL commands
# primarily deal with creating, modifying, and deleting database schema objects such as tables, indexes, and views. These
# commands operate at the schema level, which means they do not manipulate data but rather the structures that hold the data.

# Key DDL Commands:
# CREATE: Used to create new database objects such as tables, views, indexes, and databases.
CREATE TABLE Employees (
    ID INT PRIMARY KEY,
    Name VARCHAR(100),
    Salary DECIMAL(10, 2)
);


# DROP: Used to delete existing database objects, completely removing their structure and data from the database.
DROP TABLE Employees;

# ALTER: Used to modify the structure of existing database objects, such as adding or removing columns from a table.
ALTER TABLE Employees
ADD Department VARCHAR(50);

# TRUNCATE: Used to remove all data from a table, but keeps the table structure intact.
TRUNCATE TABLE Employees;
```

```python
# Q.3 What is DML ? Explain INSERT, UPDATE and DELETE with example.

# Ans.
# DML (Data Manipulation Language) is a subset of SQL used to manipulate the data within the database. While DDL deals with the
# structure and schema of the database, DML focuses on the actual data contained in the tables. Common DML operations include
# inserting, updating, and deleting records in a database.

# Key DML Commands:
# INSERT: Used to add new data (records/rows) to a table.
INSERT INTO Employees (ID, Name, Salary, Department)
VALUES (1, 'Alice Johnson', 50000, 'HR');

# UPDATE: Used to modify existing data in a table.
UPDATE Employees
SET Salary = 55000
WHERE ID = 1;

# DELETE: Used to remove data (records/rows) from a table.
DELETE FROM Employees
WHERE ID = 1;
```

```python
# Q.4 What si DQL ? Explain SELECT wtih an exapmle.

# Ans.

# DQL (Data Query Language) is a subset of SQL used for querying and retrieving data from a database. The primary operation in
# DQL is the SELECT statement, which allows users to fetch data from one or more tables based on specific conditions.

# SELECT Statement:
# The SELECT statement is used to retrieve data from a database. You can specify which columns to retrieve, apply filters with
# the WHERE clause, sort the results with ORDER BY, and more. It is one of the most widely used commands in SQL.
```

```python
# Q.5 Explain Primary Key and Foreign Key.

# Ans.
# Primary Key and Foreign Key are two important concepts in relational database design that help ensure data integrity and 
# define relationships between tables.

# 1. Primary Key:
# A Primary Key is a column (or a set of columns) in a table that uniquely identifies each row (record) in that table. It 
# enforces uniqueness and ensures that no duplicate or NULL values can exist in the column(s) marked as the primary key.

# Characteristics of a Primary Key:
# Uniqueness: Each value in the primary key column(s) must be unique across the table.
# Not NULL: Primary key fields cannot have NULL values.
# Single or Composite: A primary key can consist of one column (single primary key) or multiple columns (composite primary key).
# Table-level constraint: Every table should ideally have a primary key to uniquely identify each record.

CREATE TABLE Students (
    StudentID INT PRIMARY KEY, 
    Name VARCHAR(100),
    Age INT
);


# 2. Foreign Key:
# A Foreign Key is a column (or a set of columns) in one table that references the Primary Key in another table. It is used to
# establish and enforce a link between the data in two tables, essentially representing relationships between them.

# The foreign key enforces referential integrity between the tables, meaning that the value of the foreign key in one table 
# must match an existing value in the primary key of the referenced table.

# Characteristics of a Foreign Key:
# Link Between Tables: It establishes a relationship between two tables.
    
# Referential Integrity: The foreign key ensures that the value in the child table (the table with the foreign key) corresponds 
# to an existing value in the parent table (the table with the primary key).

# Multiple Foreign Keys: A table can have multiple foreign keys that reference the primary keys of different tables.
CREATE TABLE Customers (
    CustomerID INT PRIMARY KEY,
    Name VARCHAR(100)
);

CREATE TABLE Orders (
    OrderID INT PRIMARY KEY,
    OrderDate DATE,
    CustomerID INT,
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
);
```

```python
# Q.7 Give the order of execution of SQL clauses in an SQL queary .

# Ans.
# FROM: Specifies the source table(s) from which to retrieve or manipulate data. Joins between multiple tables are processed 
#       here.

# WHERE: Filters rows based on specified conditions before any grouping. Only rows that satisfy the WHERE condition are passed 
#        to the next stage.

# GROUP BY: Groups rows that have the same values in specified columns. This is typically used with aggregate functions 
#       (e.g., COUNT(), SUM()).

# HAVING: Filters groups of rows created by GROUP BY. It applies conditions to the aggregated data, unlike WHERE which works 
#         on individual rows.

# SELECT: Selects columns to display or calculate expressions and aggregates. This is where aliasing occurs and functions are
#         applied to the columns.

# DISTINCT: Removes duplicate rows from the result set, applied after the SELECT clause.

# ORDER BY: Sorts the result set based on one or more columns. You can specify ascending (ASC) or descending (DESC) order.

# LIMIT/OFFSET (or FETCH FIRST N ROWS ONLY): Limits the number of rows returned by the query, often used for pagination or to
#                                         restrict the result set to a specific number of rows.

SELECT DISTINCT Name, COUNT(Salary) 
FROM Employees
WHERE Salary > 50000 
GROUP BY Department 
HAVING COUNT(Salary) > 5
ORDER BY Name ASC
LIMIT 10;
```
